```python
import pennylane as qml

def qfun():
    qml.PauliZ(wires=0)
    qml.Hadamard(wires=0)
    return qml.state()

device = qml.device("default.qubit", wires=1)

qn = qml.QNode(qfun, device)
```

```python
print(qml.draw(qn)())

fig, ax = qml.draw_mpl(qn)()
```

```python
sv = qn()
print(sv)
```
